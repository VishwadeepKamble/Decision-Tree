### Dataset : load_digits 
The data set contains images of hand-written digits: 10 classes where each class refers to a digit.

#### Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
import seaborn as sns
%matplotlib inline
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

#### Load Dataset

```python
dataset = load_digits()
dataset
```

```python
df = pd.DataFrame(dataset.data, columns = dataset.feature_names)
df["target"] = dataset.target
```

```python
df.head()
```

```python
plt.gray() 
plt.matshow(dataset.images[4]) 
plt.show()
```

#### Missing Values

```python
df.isnull().sum()
```

As checked there are no missing values in dataset

#### Splitting Dependent and Independent Variables

```python
X = df.drop("target", axis = 1)
y = df["target"]
```

```python
X.shape
```

```python
y.shape
```

### Train_Test_split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state = 0)
```

### Base Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
```

```python
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
```

```python
y_train_pred = clf.predict(X_train)
y_test_pred = clf.predict(X_test)

print(classification_report(y_train, y_train_pred))
print(classification_report(y_test, y_test_pred))
```

macro avg : avg of f1-score

weighted avg : (27/360) * 0.89 and so on

#### As we can see that every value is 1.00 here, so it is clear case of over fitting, to overcome this we have to perform Gridsearch CV

### Using Gridsearch

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
from sklearn.metrics import f1_score
```

```python
param_grids = {'max_depth':np.arange(1,12),
              'criterion':['gini','entropy'],
               'min_impurity_decrease':np.arange(0.01,4),
               'ccp_alpha' :[0.001, 0.01,0.1, 1, 2,5,10]}
```

```python
clf = GridSearchCV(DecisionTreeClassifier(class_weight='balanced'), param_grid = param_grids, cv = 5, scoring = "f1_macro", n_jobs = -1)
clf.fit(X_train, y_train)

y_train_pred = clf.predict(X_train)
y_test_pred = clf.predict(X_test)

print(classification_report(y_train, y_train_pred))
print(classification_report(y_test, y_test_pred))
```

#### As we can see that the distance between test and train are now less as compared to Base Decision Tree Model
